```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from sklearn.model_selection import StratifiedKFold
```

## Pre-Processing

```python
music_df = pd.read_csv('music_genre.csv')
```

```python
# Remove columns containing irrelevant data
music_df.drop(columns=['instance_id', 'artist_name', 'track_name', 'key', 'obtained_date'], inplace=True)
```

```python
# Filter out invalid values of duration: can't be negative as it is a 
# measurement of time.
music_df = music_df[music_df['duration_ms'] > 0]
```

```python
# Turning mode values into binary values
music_df['mode'].replace('Major', 0, inplace=True)
music_df['mode'].replace('Minor', 1, inplace=True)
```

```python
music_df.head(10)
```

```python
# Defining function for removing rows containing missing values and values that will not be helpful in our analysis.
def remove_invalid_values(df):
    df.replace(np.inf, np.nan, inplace=True)
    df.replace(-np.inf, np.nan, inplace=True)
    return df.dropna(how='any', inplace=True)
```

```python
# There are some strange values in this dataset, such as '?'. We will replace 
# these values and then apply the function defined in the previous cell.
music_df.replace('?', np.nan, inplace=True)
remove_invalid_values(music_df)
```

We will reorder our attributes in X alphabetically to eliminate any confusing with encoding done by functions in our MATLAB file.

```python
music_df = music_df[['acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'loudness', 'mode', 'popularity', 'speechiness', 'tempo', 'valence', 'music_genre']]
music_df.head(10)
```

```python
# The column for the tempo variable was in string format, so we will change it to a float.
music_df = music_df.astype({'tempo': 'float64'})
```

```python
# Change the unit of measurement for the duration column from milliseconds to seconds.
music_df['duration_ms'] = music_df['duration_ms'].div(1000)
music_df.rename(columns={'duration_ms':'duration_s'}, inplace=True)
```

## EDA of Target Class

```python
# Assigning dependent variable, Y, for distribution analysis.
Y = music_df['music_genre']
```

```python
# Histogram of counts for each different possible target variable value.
counts, bins, patches = plt.hist(Y, rwidth=0.85)
plt.ylabel('Count')
plt.xlabel('Target Variables')
plt.rc('xtick', labelsize=5)
```

```python
# Acquiring the counts for each genre
genres = ['Electronic', 'Anime', 'Jazz', 'Alternative', 'Country', 'Rap', 'Blues', 'Rock', 'Classical', 'Hip-Hop']
for i in range(len(counts)):
    print("The data contains {} instances of the genre {}.".format(counts[i], genres[i]))
```

Our target variable is very close to being evenly distributed.

We have many different target variable classes, with a few that are very similar in terms of musical features: Rap and Hip-Hop, for example. This could create complications in our model fitting, so we will take 6 of the most popular genres, listed in the dataset, that are adequately distinct from each other.

```python
keep_genres = ['Electronic', 'Jazz', 'Alternative', 'Blues', 'Classical', 'Hip-Hop']
max_samples_per_genre = 4000
reduced_music_df = pd.DataFrame(columns=music_df.columns)
for genre in keep_genres:
    genre_samples = music_df[music_df['music_genre'] == genre].sample(n = max_samples_per_genre)
    reduced_music_df = reduced_music_df.append(genre_samples)
```

A maximum number of 4000 observations is set for each of the 6 genres we decided to retain in our dataset. This was due to the dataset being too large for the algorithms to run in a reasonable amount of time. 4000 was decided upon as it maximised our dataset, whilst making sure the algorithms didn't take an excessive amount of computation time.

## Pearson Correlation Coefficient Analysis for Independent Variables

```python
# Assigning independent variables, X, from our dataset in order to check the
# features for multicollinearity.
X = reduced_music_df.drop(columns=['music_genre'])
```

```python
# Calculating pearson correlation matrix for the independent variables.
pearson_matrix = X.corr(method='pearson')
sns.set(font_scale=0.75)
sns.heatmap(pearson_matrix, cmap='PuOr', vmin=-1, vmax=1, annot = True)
plt.title('Pearson Correlation Matrix for Features')
plt.savefig('EDA_graphs/feature_corr_matrix.svg', bbox_inches='tight')
plt.show()
```

We can see, from the above matrix, that energy and loudness have a very high correlation coefficient of 0.84. We also see high coefficients between loudness and acousticness, -0.73, as well as energy and acousticness, -0.79. We can import this dataset into Matlab and calculate the Variance Inflation Factor of each of the features in X. This will inform which of these features should be omitted from our logistic regression.
